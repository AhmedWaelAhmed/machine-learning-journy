list ,set dic with for loops

```python
square =[]
for i in range(11):
    
    if (i%2==0):
        square.append(i**2)
print(square)
#list comprehension
squares =[i**2 for i in range(11) if i%2==0]
print(squares)
# if want to make it set use {}
squares ={i**2 for i in range(11) if i%2==0}
print(type(squares))
#dic comprehension
squares_dic ={i:i**2 for i in range(11) if i%2==0}
print(squares_dic)

# matrix using for loop

matrix =[]
for i in range(3):
    row =[]
    for j in range(4):
        row.append(j)
    matrix.append(row)
print(matrix)
```

```python
# old way to create class
class product:
    pass
appl14= product()
ultra12= product()
print(appl14)
print(ultra12)
# add acttributes to class
appl14.name="iphone 14"
appl14.price=1200
ultra12.name="samsung ultra 12"
ultra12.price=1100
print(appl14.name, appl14.price)
print(ultra12.name, ultra12.price)


# using constructor new way
class product:
    #class attributes
    type= "mobile" # all object will share this attribute like iphone and samsung
    def __init__(self, namees, price, count): # constructor  
        self.namee = namees #set attributes instance variables   
        self.price = price #set attributes
        self.count= count
iphone= product(namees="iphone 14", price=1200, count=5)
samsung= product("samsung ultra 12", 1100, 3)
print(iphone.namee, samsung.namee)
print(iphone.price, samsung.price)
print(iphone.count, samsung.count)
print(iphone.type, samsung.type)
# making methods in class
def discount(self,discoun_ratio):
    self.price= self.price - (self.price * discoun_ratio)
print(iphone.price)
discount(iphone,0.1)
print(iphone.price)

```
