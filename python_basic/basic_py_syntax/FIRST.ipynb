encapsulation for class :التغليف  to protection 
getters 
setters




```python
class Product:
    def __init__(self, ids, names, price):
        self.id = ids       # Public attribute
        self.name = names   # Public attribute
        self.__price = price # Private attribute (Protected by __)

    def discount(self, discount_ratio):
        # This method belongs to the class, so it CAN access __price
        self.__price = self.__price - (self.__price * discount_ratio)
        
    def info(self):
        # This method belongs to the class, so it CAN read __price
        return f"Product ID: {self.id}, Name: {self.name}, Price: {self.__price}" 
        
    def set_price(self, new_price):
        # This is a Setter Method. It allows controlled access to change the price.
            self.__price = new_price
        

# --- TESTING THE CODE ---

p1 = Product(1, "iPhone 14", 1200)

print("--- 1. Initial State ---")
print(p1.info())  # Shows 1200

print("\n--- 2. Trying to change price directly (The Mistake) ---")
p1.__price = 100  # This looks like it works, but it DOES NOT change the real price!
print(f"You tried to set p1.__price to: {p1.__price}") 
print(f"BUT the real internal price is still: {p1.info()}")

print("\n--- 3. Using the Setter Method (The Correct Way) ---")
p1.set_price(1000) # We ask the class to change the price for us
print(p1.info())   # Now it shows 1000
```

derived class inherte from base class (inheratence)
father name is super 

```python
class Product:
    def __init__(self, id, name, price, count):
        self.id = id
        self.name = name
        self._price = price  # لاحظ استخدام _price هنا
        self.count = count

    def info(self):
        return f'id: {self.id}, Name: {self.name}, Price: {self._price}$, Count: {self.count}'

class Mobile(Product):
    def __init__(self, id, name, price, count, screen_size, camera):
        # نستخدم super لإرسال البيانات الأساسية للكلاس الأب
        super().__init__(id, name, price, count)
        
        # نعرف فقط المتغيرات الجديدة الخاصة بالموبايل
        self.screen_size = screen_size
        self.camera = camera

    # نقوم بتعديل دالة info لإضافة تفاصيل الموبايل
    def info(self):
        # نستدعي info القديمة ونضيف عليها البيانات الجديدة
        return f'{super().info()}, Screen: {self.screen_size}, Camera: {self.camera}'

class Laptop(Product):
    def __init__(self, id, name, price, count, screen_size, RAM):
        # نستخدم super هنا أيضاً
        super().__init__(id, name, price, count)
        
        self.screen_size = screen_size
        self.RAM = RAM

    def info(self):
        return f'{super().info()}, Screen: {self.screen_size}, RAM: {self.RAM}'


# --- التجربة ---
samsungA12 = Mobile(1, 'Samsung A12', 300, 5, '6.5 inch', '48MP')
MACBOOK = Laptop(2, 'Macbook Air', 1500, 3, '13 inch', '8GB')

print(MACBOOK.info())
print(samsungA12.info())
```

Abstraction 
اخفاء التفاصيل     

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod 
    def area(self):  # must be implemented by subclasses
        pass
         
class Rectangle(Shape):
    # Proper constructor and area implementation
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)
class sqare(Shape):
    def __init__(self,s):
        self.s=s
    def area(self):
        return self.s *self.s

        

# create a Rectangle with width=3 and height=4, then print its area
print(Rectangle(3, 4).area())
print(Circle(5).area())
print(sqare(5).area())
# The Shape class is now abstract and cannot be instantiated directly.
# Only Rectangle and Circle can be instantiated, and they must implement the area method.
```
