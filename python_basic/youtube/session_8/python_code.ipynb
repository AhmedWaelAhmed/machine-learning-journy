```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary  # Private (Locked)

    # Getter: To READ the salary safely
    def get_salary(self):
        return self.__salary

    # Setter: To UPDATE the salary with Validation
    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary
            print(f"Salary updated to: {self.__salary}")
        else:
            print("Error: Salary cannot be negative!")

emp = Employee("Depi", 1000)
```

```python
# 1. Try to access directly -> Fails
#print(emp.__salary)  # Error!

# 2. Use Getter -> Works
print(emp.get_salary())  # Output: 1000

# 3. Use Setter (Valid logic) -> Works
emp.set_salary(5000)     # Output: Salary updated to: 5000
print(emp.get_salary())
# 4. Use Setter (Invalid logic - Negative Number) -> Blocked
emp.set_salary(-200)     # Output: Error: Salary cannot be negative!
```

```python
class A:
    def do_this(self): print("I am in A")
class B(A): pass
class C:
    def do_this(self): print("I am in C")

# D inherits from B first, then C
class D(B, C): pass

d = D()
d.do_this() # Output: "I am in A"
```
